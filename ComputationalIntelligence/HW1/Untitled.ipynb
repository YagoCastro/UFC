### 2.3 Medidas de Posição

Que são para variáveis quantitativas
- Média
- Mediana
- Moda
- Percentis
- Quartis
- Decis


Média

É média aritmética simples para a variável, ou seja, é a soma dos valores (Xi) contidos na variável dividido pela quantidade total de observações (n)

A média aritmética é dada pela seguinte fórmula:

$$
\bar{X} = \frac{\sum_{i=1}^{n}X_i }{n}
$$

$$
\sum_{i=1}^{n} x_i = x_1 + x_2 + x_3 + \dots + x_n
$$

onde:

- $\\bar{X}\$ é a média aritmética,

- \(x_1, x_2, x_3, \dots, x_n\) são os valores observados,
- \(n\) é o número total de valores.


Mediana

É o elemento cental da distribuição da variável, considerando que a variável esteja com seus n valores organizados de forma crescente

Metade dos valores da variável são mariores ou iguais ao valor da mediana e metade dos valores são menores ou iguais ao valor da mediana 

$$
Md(X) = 
\begin{cases} 
\frac{X_{\frac{n}{2}} + X_{\left(\frac{n}{2} + 1\right)}}{2}, & \text{se } n \text{ for par} \\
X_{\left(\frac{n+1}{2}\right)}, & \text{se } n \text{ for ímpar}
\end{cases}
$$


Moda

- É o valor que aparece com maior frequência nas observações de uma variável
- A moda também pode ser valculada para dados qualitativos
- É possível que não exista a moda de uma variável (especialmente, se for uma variável contínua)

- Ocorre quando nenhum valor se repete
- Quando Média = Mediana = Moda, caracteriza uma distribuição simétrica

Percentis

São os elementos da distribuição da variável que dividem as observações em cem partes iguais, considerando que a variável esteja com seus valores organizados de forma crescente

- 14° Percentil
- 42° Percentil
- 60° Percentil ...

$$
Pos (P_i) = [(n-1) \cdot \left(\frac{P_i}{100}\right)]+1
$$

Quartis

São elementos da distribuição da variável que dividem as observações em quatro partes iguais, considerando que que a variável esteja com seus valores organizados de forma crescente

- 1° Quartil: 25% das observações são menores do que o 1° quartil
- 2° Quartil: trata-se da mediana
- 3° Quartil: 25% das observações são maiores do que o 3° quartil

- 1° Quartil = 25° Percentil
- 2° Quartil = 50° Percentil
- 3° Quartil = 75° Percentil


Decis

São os elementos da distribuição da variável que dividem as observações em dez partes iguais, considerando que a variável esteja com seus valores organizados de forma crescente.

1° Decil
2° Decil
8° Decil

1° Decil = 10° Percentil
3° Decil = 30° Percentil
8° Decil = 80° Percentil

### 2.4 Medidas de Dispersão

- Amplitude
- Variância
- Desvio padrão
- Erro padrão
- Coeficiente de variação

Amplitude

Apresenta a diferença entre o valor máximo e o valor mínimo de uma variável

$$
A = X_{máx} - X_{min}
$$

- Valor máximo: maior valor da variável
- Valor mínimo: menor valor da variável


Variância

Mostra a dispersão das observações de uma variável em torno de sua média

$$
S^2 = \frac{\sum_{i=1}^{n}(X_i - \bar{X})^2 }{n - 1}
$$

Neste caso, trata-se da variância amostral

Desvio padrão

É uma medida derivada da variância, tornando mais simples a interpretação da dispersão em torno da média

- A variância é definida em termos quadrados, o que dificulta a interpretação

O desvio padrão é a raiz quadrada da variância

$$
S = \sqrt{S^2}
$$

Um alto desvio padrão significa que os dados estão muito dispersos em relação a média

Erro padrão

É o desvio padrão da média da variável

$$
S_{\bar{X} = \frac{S}{\sqrt{n}}
$$

Sendo que S é o desvio padrão da variável e n o tamanho da amostra

Quanto maior o tamanho da amostra, menor o erro padrão na estimativa da média da variável -> mais precisa é a média estimada

Coeficiente de variação (CV)

É uma medida de dispersão relativa, pois relaciona o desvio padrão e a média de uma variável

- Pode ser utilizada para realizar comparações entre amostras, por exemplo
- Quanto menor o CV, mais homogêneos são os valores da variável e mais concentados estão os valores em torno da média

$$
CV = \frac{S}{X} \cdot 100 
$$

### 2.5 Medias de forma

- Assimetria 
- Curtose

Assimetria: local de concentração da distribuição

Curva Simétrica: Média = Mediana = Moda
Curvas Assimétricas - Direita: tem cauda mais longa à direita -> Média > Mediana
Curvas Assimétricas - Esquerda: tem cauda mais longa à esquerda -> Média < Mediana

Coeficiente de Assimetria de Fisher

$$
g_{1} = \frac{n^2 \cdot M_3}{(n-1) \cdot (n-2) \cdot S^3}
$$
 em que 
$$
M_{3} = \frac{\sum_{i=1}^{n}(X_i - \bar{X})^3 }{n}
$$

Assimetria:
g1 = 0 indica curva simétrica
g1 > 0 indica curva assimétrica positiva (à direita)
g1 < 0 indica curva assimétricaca negativa (à esquerda)

Curtose: Achatamento da curva de distribuição

Com a curva normal como referência, é possível observar se as curvas têm menor dispersão em torno da média ou maior disper~sao em torno da média.

Coeficiente de curtose de Fisher:

$$
g_{2} = \frac{n^2 \cdot (n+1) \cdot M_4}{(n-1) \cdot (n-2) \cdot (n-3) \cdot S^4} - 3 \cdot \frac{(n-1)^2}{(n-2) \cdot (n-3)}
$$
 em que 
$$
M_{3} = \frac{\sum_{i=1}^{n}(X_i - \bar{X})^4 }{n}
$$

g2 = 0 indica curva com distribuição normal
g2 > 0 indica curva com distribuição alongada
g2 < 0 indica curva com distribuição achatada

## Relação Entre Variáveis

Medidas de análses bivariadas

### Teste qui-quadrado: variáveis qualitativas

Inicia-se por meio de uma tabela de distribuição conjunta de frequênqua, a tabela de contigência (ou tabela de classificação cruzada) que apresenta as frequências absolutas observadas para cada par de categorias das variáveis

Em seguida, saõ identificadas as frequências absolutas esperadas para cada par de categorias das variáveis

$$
    freq.absoluta esperada_11 = \frac{(\sum {L1} \cdot \sum {C1})}{N}
$$

O mesmo cálculo é realizado para cada par de categorias da tabela de contingência, mantendo-se as respectivas correspond^neicas de linha e coluna no numerador

Posteriormente, são identificados os resíduos para cada par de categorias das variáveis

$$
resíduo_{11} = freq.absoluta observado_{11} - freq. absoluta esperada_{11}
$$

Os resíduos são identificados para cada par de categorias da tabela de contingência

Por fim, são calculados os valores X^2 individuais de cada par de categorias

$$
X^2 = \frac{(resíduo)^2}{(freq.absoluta esperada)}
$$

E são somados valores $X^2$ individuais para obter o valor X^2 total da análise

Com base no valor X^2 total, realiza-se o seguinte teste:

- $H_{0}:$ As variáveis se associam de forma aleatória
- $H_{1}:$ A associação entre as variáveis não se dá de forma aleatória

Dados o nível de significância e os graus de liberdade, se o valor da estatística $X^2$ for maior do que seu valor crítico, há associação significante entre as duas variáveis $(H_1)$

Valor crítico da distribuição $X^2$ com $(l-1) \cdot (J-1)$ graus de liberdade

A distribuição X^2 apresenta curva assimétrica e positiva

É formada a partir da soma dos quadrados de v variáveis aleatórias independentes com distribuição normal padrão

Exemplo: Valores críticos em uma tabela qui-quadrado e no Excel.

- região de rejeição da hipótese nula

Uma possivel conclusão: Como o valor calculado para a estatística qui^2 é maior do que o valor crítico ao nível de significância de 5%, rejeita-se H0. Portante, as variáveis são associadas.

p-valor < nível de significância (5%)

### 2.3.4 Coeficiente de correlação de Pearson

É utilizado para identificar a correlação entre duas variáveis quantitativas

Inicia-se pelo cálculo da covariância entre as duas variáveis e, posteriormente, obtém-se o coeficiente de correlação de Pearson

$$
cov(X, Y) = \frac{\sum_{i=1}^{n} (X_{i} - \bar{X}) \cdot (Y_{i} - \bar{Y})}{n - 1}
$$

$$
p_{XY} = \frac{cov(X, Y) }{S_{X} \cdot {S_y}}
$$
, sendo que $p_{XY}$ varia entre -1 e 1


$$
- p_{XY} = -1 (perfeita negativa)
- p_{XY} = 0  (sem correlação)
- p_{XY} =  1 (perfeita positiva)
$$ 

## Variáveis e as distribuições

## Distribuições de Probabilidade

### Distribuição uniforme discreta

Todos os possíveis valores têm a mesma probabilidade de ocorrência 

$$
P(X = x_i) = \frac{1}{n}
$$

O parâmetro n representa a quantidade de possíveis valores

Exemplo: As probabilidades dos resultados possíveis ao lançar um dado são:

$$
- P(X=1) = 1/6
- P(X=1) = 1/6
- P(X=2) = 1/6
- P(X=3) = 1/6
- P(X=4) = 1/6
- P(X=5) = 1/6
- P(X=6) = 1/6
$$

### Distribuição de Bernoulli 'Logística Binária'

Os valores da variável podem assumir apenas dois resultados possíveis, sendo que tais resultados são chamados de sucesso (x=1) ou fracasso (x=0)

A distribuição de Bernoulli apresenta a probabilidade de sucesso (p) ou de fracasso (1-p) quando ocorre apenas um experimento

$$
P(X = x) = p^x \cdot (1-p)^(1-x)
$$

Exemplo: A probabilidade (p) de que um candidato seja aprovado (x = 1) em um exame para um conselho de classes é de 48%. Se cada candidato só pode realizar o exame uma vez, analise as probabilidades possiveis por meio da distribuilção de Bernoulli

- $P(X=1) = (0,48)^1 \cdot (1 - 0,48)^0 = 0,48 ou 48%$
- $P(X=0) = (0,48)^0 \cdot (1 - 0,48)^1 = 0,52 ou 52%$

Sendo que X = 1 é aprovação e X = 0 é a reprovação no exame.

### Distribuição de Binomial 'Logística Multinomial!'

A distribuição binomial ocorre quando há (n) repetições independentes do experimento de Bernoulli e a probalidade de sucesso (p) é constante em todas as repetições

A variável no modelo binominal indica a quantidade de sucessos (k) nas (n) repetições do experimento

$$ P(X = k) = \binom{n}{k} \cdot p^k \cdot (1 - p)^{n - k} \quad \text{em que} \quad \binom{n}{k} = \frac{n!}{k!(n - k)!} $$


### Distribuição de Binomial Negativa 

Na distribuição binomial negativa, são realizados (x) ensaios independentes de Bernoulli até que sejam obtidos (k) sucessos

A probabilidade de sucesso (p) é constante em todos os ensaios realizados

A variável no modelo binominal negativa indica quantidade de ensaios (x)

$$ P(X = k) = \binom{n-1}{k-1} \cdot p^k \cdot (1 - p)^{x - k} \quad \text{em que} \quad \binom{n-1}{k-1} = \frac{(x-1)!}{(k-1)![((x-1) - (k-1))]!} $$


### Distribuição de Poisson

A distribuição Poisson indica probabilidade do número de sucessos (k) em uma determinada exposição contínua

Exemplos de exposição: tempo e área

$$
P(X=k) = \frac{e^{-\lambda} \cdot \lambda^{k}}{k!}
$$

em que lambda é a taca média estimada de ocorrencia do exento (sucesso) em dada exposição

### Distribuição Normal

É a distribuição Gaussina, com curva em forma de sino

Os parâmetros relevantes da distribuição normal são a média (mi) e o desvio padrão($\sigma$) da variável

$$
f(x) = \frac{1}{\sigma \cdot \sqrt{2 \cdot \pi}} \cdot e^{- \frac{(x - μ)^2}{2 \cdot \sigma^2}}
$$

### Distribuição Normal Padrão

Para obter a normal padrão, transforma-se a variável por meio do Z-score

A transformação pelo Z-score faz com que a variável passe a ter média =0 e desvio padrão = 1 e não altera a distribulão original

$$
Z = \frac{(X - μ)}{\sigma}
$$
