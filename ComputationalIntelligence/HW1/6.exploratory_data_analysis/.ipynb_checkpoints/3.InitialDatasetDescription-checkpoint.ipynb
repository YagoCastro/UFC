### 2.3 Qual a variável Alvo?

Assim como o relatório Seattle Energy Benchmarking Analysis 2016 que visa analisar o consumo energético e as emissões de gases de efeito estufa em prédios de Seattle, com foco nos dados coletados durante o ano de 2016. O objetivo principal é entender o desempenho energético dos edifícios da cidade, identificar oportunidades de redução de consumo e de emissões, e fornecer insights para a formulação de políticas de eficiência energética. Além disso, o relatório avalia a evolução das pontuações ENERGY STAR e os progressos em eficiência energética, e serve como um recurso para guiar futuras melhorias e regulamentações relacionadas ao consumo e às emissões nos prédios da cidade.

Sendo assim será usado como variável alvo ou variável preditiva a variável ENERGYSTARScore, que é uma métrica desenvolvida pela EPA (Environmental Protection Agency) dos Estados Unidos para avaliar a eficiência energética de edifícios comerciais e residenciais em comparação com edifícios semelhantes. Esse escore vai de 1 a 100 e é calculado com base no consumo de energia real do edifício, ajustado para fatores como clima, ocupação e uso específico de cada tipo de construção. Essa variável resume o desempenho energético geral de um edifício em uma única métrica, considerando diversos fatores subjacentes.
Embora a pontuação seja contínua, para o caso aqui de estudo e análise será criado classes que representam a eficiência do edifício, essa classificação ajuda a distinguir melhor os edifícios que precisam de mais atenção, facilitando a priorização de intervenções de eficiência energética. As categorias são: 

- 0 a 25: Muito ineficiente
- 26 a 49: Ineficiente
- 50 a 74: Moderadamente eficiente
- 75 a 100: Altamente eficiente

### Detalhamento da pontuação

Muito Ineficiente (0 a 25): Esses edifícios estão entre os 25% com menor eficiência energética, indicando um alto potencial para melhorias significativas. Edifícios nessa faixa podem ter sistemas de energia ultrapassados, alta demanda de energia ou baixa manutenção.

Ineficiente (26 a 49): Edifícios nessa categoria têm um desempenho inferior à média, mas podem ser menos críticos que os da faixa “Muito Ineficiente”. No entanto, ainda existe potencial para intervenções que melhorem o consumo de energia.

Moderadamente Eficiente (50 a 74): Estes edifícios apresentam um desempenho energético médio, com algum nível de eficiência. Embora ainda exista espaço para melhorias, eles já estão dentro de uma faixa de eficiência razoável.

Altamente Eficiente (75 a 100): Edifícios nessa categoria estão entre os mais eficientes, indicando baixo consumo de energia em comparação com o uso típico do edifício. Esses podem ser candidatos para certificações e são modelos de sustentabilidade.

### Como Funciona o ENERGY STAR Score:

Escala de 1 a 100: Um escore de 50 indica desempenho energético mediano, ou seja, o edifício está na média do consumo de energia para o seu tipo. Um escore acima de 75 significa que o edifício está entre os 25% mais eficientes da categoria e pode receber a certificação ENERGY STAR.
Ajuste por Fatores Externos: O escore é ajustado para levar em conta diferenças de clima, tamanho, tipo de uso do edifício e outras condições operacionais para garantir que a comparação seja justa entre edifícios diferentes.
Consumo Real de Energia: Baseia-se no consumo anual de energia do edifício em comparação com dados de outros edifícios do mesmo tipo, como escolas, escritórios, hospitais, etc.

### Por que o ENERGY STAR Score é Útil?

Comparação de Desempenho: Ele permite que gestores e proprietários comparem o desempenho energético de seus edifícios com benchmarks nacionais, ajudando a identificar edifícios eficientes e aqueles que podem se beneficiar de melhorias.
Redução de Custos e Sustentabilidade: Um escore alto geralmente indica menor consumo de energia e, portanto, menores custos de operação e menor impacto ambiental.
Certificação e Reconhecimento: Edifícios com pontuação de 75 ou mais podem solicitar a certificação ENERGY STAR, um selo de reconhecimento de eficiência energética nos EUA.

### Fatores que Influenciam o ENERGY STAR Score

O cálculo do escore ENERGY STAR utiliza vários fatores:

- Consumo de Energia: Total de energia utilizada (por exemplo, kWh ou kBtu).
- Área do Edifício: Dimensões do edifício ou área total útil.
- Uso do Edifício: Cada tipo de edifício tem um perfil energético típico, e o escore considera essas diferenças.
- Clima e Localização: Dados como Heating Degree Days (HDD) e Cooling Degree Days (CDD) são usados para ajustar o escore para regiões com condições climáticas diversas.
- Ocupação e Equipamentos: Edifícios com alta densidade de ocupação ou com muitos equipamentos podem ter consumo elevado, e o escore é ajustado para considerar isso.

### Como Calcular ou Obter o ENERGY STAR Score

Embora o cálculo exato seja um segredo comercial, os dados necessários para a análise estão disponíveis em plataformas como o ENERGY STAR Portfolio Manager, que calcula o escore automaticamente após inserir as informações relevantes do edifício.

Ter um ENERGY STAR Score alto é um indicador positivo da eficiência energética do edifício e pode ser um diferencial importante para certificações ambientais, marketing e reconhecimento de sustentabilidade.

https://www.energystar.gov/buildings/benchmark/understand-metrics/how-score-calculated

```python
import pandas as pd

# Código de Cores ANSI
azul = "\033[38;5;33m"  #
verde = "\033[38;5;82m"  #
laranja = "\033[38;5;208m"  
reset = '\033[0m'

def calcular_estatisticas(df, coluna):
    """
    Calcula estatísticas descritivas de uma coluna numérica de um DataFrame.

    Parameters:
    df (pd.DataFrame): O DataFrame contendo os dados.
    coluna (str): O nome da coluna para calcular as estatísticas.

    Returns:
    dict: Um dicionário contendo várias estatísticas, ou uma mensagem de erro se a coluna não for numérica.
    """
    if coluna not in df.columns:
        return {"erro": "A coluna especificada não existe no DataFrame."}

    if not pd.api.types.is_numeric_dtype(df[coluna]):
        return {"erro": "A coluna especificada não é numérica."}
    # INFORMAÇÕES SOBRE O DATASET
    print('----------- Informações sobre o Dataset -----------')
    num_linhas = df.shape[0]
    print(f'Número de linhas do dataset: {azul}{num_linhas}{reset}')
    
    # Calculando a quantidade de valores únicos
    valores_unicos = df[coluna].nunique()
    print(f'Valores únicos da coluna: {azul}{valores_unicos}{reset}')
    
    # MEDIDAS DE POSIÇÃO
    print('----------- Medidas de Posição -----------')
    
    # Calculando a Média
    media = df[coluna].mean()
    print(f'média: {azul}{media}{reset}')
    
    # Calculando a Mediana
    mediana = df[coluna].median()
    print(f'mediana: {azul}{mediana}{reset}')
    
    # Calculando a Moda
    frequencia = df[coluna].value_counts()
    max_frequencia = frequencia.max()
    modas = frequencia[frequencia == max_frequencia].index.tolist()
    
    if max_frequencia == 1:
        print("Não há moda.")
    elif len(modas) == len(frequencia):
        print("Não há moda, todos os elementos aparecem com a mesma frequência.")
    else:
        print(f'moda: {azul}{modas}{reset}')
        print(f"Frequência da moda: {max_frequencia}")
    
    # Calculando os Quartis
    quartis = df[coluna].quantile([0.25, 0.5, 0.75]).tolist()  # Q1, Q2 (mediana), Q3

    # Exibindo os Quartis
    print(f'Primeiro Quartil (Q1) [25%]: {azul}{quartis[0]}{reset}')
    print(f'Segundo Quartil (Mediana, Q2) [50%]: {azul}{quartis[1]}{reset}')
    print(f'Terceiro Quartil (Q3) [75%]: {azul}{quartis[2]}{reset}')
    
    # MEDIDAS DE Dispersão
    print('----------- Medidas de Dispersão -----------')
    
    # Calculando o Máximo do conjunto
    maximo  = df[coluna].max()
    print(f'O máximo do conjunto é  {azul}{maximo}{reset}')
    
    # Calculando o Mínimo do conjunto
    minimo  = df[coluna].min()
    print(f'O minimo do conjunto é  {azul}{minimo}{reset}')
    
    # Calculando a Amplitude
    amplitude  =  df[coluna].max() - df[coluna].min()
    print(f'A amplitude do conjunto é  {azul}{amplitude}{reset}')
    
    # Calculando a Variância
    variancia = df[coluna].var()
    print(f'A a variância do conjunto: {azul}{variancia}{reset}')
    
    # Calculando a Desvio
    desvio = df[coluna].std()
    print(f'O desvio padrão do conjunto: {azul}{desvio}{reset}')
    
    # Calculando o Erro padrão
    erro_padrao = desvio / (len(df[coluna]) ** 0.5)  # Erro padrão da média
    print(f'O Erro padrão: {azul}{erro_padrao}{reset}')
    
    # Calculando o Coeficiente de variação
    coef_variacao = (desvio / media) * 100 if media != 0 else None  # Coeficiente de variação
    print(f'O Coeficiente de variação: {azul}{coef_variacao}{reset}')
    
    # Calculando a Assimetria
    assimetria = df[coluna].skew()  # Assimetria
    print(f'A assimetria: {azul}{assimetria}{reset}')
    
    # Calculando a Curtose
    curtose = df[coluna].kurtosis()  # Curtose
    print(f'A curtose: {azul}{curtose:.2f}{reset}')

    return {
    }

# Exemplo de uso
# df = pd.DataFrame({'coluna_exemplo': [1, 2, 3, 4, 5]})
# resultado = calcular_estatisticas(df, 'coluna_exemplo')
# print(resultado)
```

```python
# Função para atualizar o gráfico com o número de bins variável e bordas nas colunas
def update_histogram(df, column, bins, title_x):
    # Cálculo das estatísticas para a coluna especificada
    mean_value = df[column].mean()
    median_value = df[column].median()
    q1_value = df[column].quantile(0.25)
    q3_value = df[column].quantile(0.75)
    
    # Criando o histograma com bordas nas colunas
    fig = px.histogram(df, x=column, nbins=bins, title=f"Histograma de {column}")
    fig.update_traces(marker=dict(line=dict(color='black', width=1)))  # Adicionando bordas pretas nas colunas
    fig.update_layout(xaxis_title=f"{title_x}", yaxis_title="Frequência", bargap=0.1)
    
    # Adicionando linhas para média, mediana, Q1 e Q3
    fig.add_shape(type="line", x0=mean_value, x1=mean_value, y0=0, y1=1, line=dict(color="Red", width=2), xref="x", yref="paper")
    fig.add_shape(type="line", x0=median_value, x1=median_value, y0=0, y1=1, line=dict(color="Green", width=2), xref="x", yref="paper")
    fig.add_shape(type="line", x0=q1_value, x1=q1_value, y0=0, y1=1, line=dict(color="Blue", width=2), xref="x", yref="paper")
    fig.add_shape(type="line", x0=q3_value, x1=q3_value, y0=0, y1=1, line=dict(color="Purple", width=2), xref="x", yref="paper")
    
    # Adicionando anotações
    fig.add_annotation(x=mean_value, y=0.95, xref="x", yref="paper", text=f"Média = {mean_value:.2f}", showarrow=False, font=dict(color="Red"))
    fig.add_annotation(x=median_value, y=0.90, xref="x", yref="paper", text=f"Mediana = {median_value:.2f}", showarrow=False, font=dict(color="Green"))
    fig.add_annotation(x=q1_value, y=0.85, xref="x", yref="paper", text=f"Q1 = {q1_value:.2f}", showarrow=False, font=dict(color="Blue"))
    fig.add_annotation(x=q3_value, y=0.80, xref="x", yref="paper", text=f"Q3 = {q3_value:.2f}", showarrow=False, font=dict(color="Purple"))
    
    # Exibindo o gráfico
    fig.show()

# Função para atualizar o gráfico interativamente com o slider
def interactive_histogram(df, column, title_x):
    # Slider para o número de bins
    bins_slider = widgets.IntSlider(value=20, min=2, max=100, step=2, description='Bins:')
    output = widgets.interactive_output(update_histogram, {'df': widgets.fixed(df),
                                                           'column': widgets.fixed(column),
                                                           'title_x': widgets.fixed(title_x),
                                                           'bins': bins_slider})
    
    # Exibindo o slider e o gráfico
    display(bins_slider, output)
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display

# Função para atualizar o gráfico com o número de bins variável e bordas nas colunas
def update_histogram(df, column, bins, title_x):
    # Cálculo das estatísticas para a coluna especificada
    mean_value = df[column].mean()
    median_value = df[column].median()
    q1_value = df[column].quantile(0.25)
    q3_value = df[column].quantile(0.75)
    
    # Criando uma figura com 3 linhas e 1 coluna
    fig = make_subplots(rows=3, cols=1, 
                        subplot_titles=("Gráfico 1: YearBuilt", f"Histograma de {column}", "Gráfico 3: Boxplot YearBuilt"))

    # Gráfico 1: YearBuilt (Gráfico de linha)
    tabela_frequencias = df['YearBuilt'].value_counts().reset_index()
    tabela_frequencias.columns = ['Ano de Construção', 'Frequência']
    tabela_frequencias = tabela_frequencias.sort_values(by='Ano de Construção')
    
    fig.add_trace(go.Scatter(x=tabela_frequencias['Ano de Construção'], 
                             y=tabela_frequencias['Frequência'], 
                             mode='lines', 
                             name='Frequência YearBuilt'), row=1, col=1)

    # Gráfico 2: Histograma
    fig.add_trace(go.Histogram(x=df[column], nbinsx=bins, name=f'Histograma de {column}', 
                               marker=dict(line=dict(color='black', width=1))), row=2, col=1)
    
    # Adicionando linhas para média, mediana, Q1 e Q3 no histograma
    fig.add_shape(type="line", x0=mean_value, x1=mean_value, y0=0, y1=1, line=dict(color="Red", width=2), xref="x2", yref="paper")
    fig.add_shape(type="line", x0=median_value, x1=median_value, y0=0, y1=1, line=dict(color="Green", width=2), xref="x2", yref="paper")
    fig.add_shape(type="line", x0=q1_value, x1=q1_value, y0=0, y1=1, line=dict(color="Blue", width=2), xref="x2", yref="paper")
    fig.add_shape(type="line", x0=q3_value, x1=q3_value, y0=0, y1=1, line=dict(color="Purple", width=2), xref="x2", yref="paper")
    
    # Adicionando anotações
    fig.add_annotation(x=mean_value, y=0.95, xref="x2", yref="paper", text=f"Média = {mean_value:.2f}", showarrow=False, font=dict(color="Red"))
    fig.add_annotation(x=median_value, y=0.90, xref="x2", yref="paper", text=f"Mediana = {median_value:.2f}", showarrow=False, font=dict(color="Green"))
    fig.add_annotation(x=q1_value, y=0.85, xref="x2", yref="paper", text=f"Q1 = {q1_value:.2f}", showarrow=False, font=dict(color="Blue"))
    fig.add_annotation(x=q3_value, y=0.80, xref="x2", yref="paper", text=f"Q3 = {q3_value:.2f}", showarrow=False, font=dict(color="Purple"))
    
    # Gráfico 3: Boxplot da variável YearBuilt
    fig_box = px.box(df, y='YearBuilt', title="Boxplot da Variável 'YearBuilt'", labels={'YearBuilt': 'Ano de Construção'})
    fig.add_trace(fig_box.data[0], row=3, col=1)

    # Atualizando o layout
    fig.update_layout(
        height=1000, width=800,
        showlegend=False,
        title_text="Análise de YearBuilt com Histograma e Boxplot",
        xaxis_title="Ano de Construção", 
        yaxis_title="Frequência",
        xaxis2_title=title_x, 
        yaxis2_title="Frequência",
        bargap=0.1
    )
    
    # Exibindo o gráfico
    fig.show()

# Função para atualizar o gráfico interativamente com o slider
def interactive_histogram(df, column, title_x):
    # Slider para o número de bins
    bins_slider = widgets.IntSlider(value=20, min=2, max=100, step=2, description='Bins:')
    output = widgets.interactive_output(update_histogram, {'df': widgets.fixed(df),
                                                           'column': widgets.fixed(column),
                                                           'title_x': widgets.fixed(title_x),
                                                           'bins': bins_slider})
    
    # Exibindo o slider e o gráfico
    display(bins_slider, output)
```

```python
interactive_histogram(df3_raw_seattle, 'YearBuilt', 'Ano de Construção')
```
