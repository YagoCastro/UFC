```python
# Filtra apenas os anos que voc√™ quer verificar
df_filtered = df[df['DataYear'].isin([2020, 2021, 2022])]

# Agrupa o DataFrame pelo nome do edif√≠cio e conta o n√∫mero de anos √∫nicos presentes para cada um
year_counts = (
    df_filtered.groupby('BuildingName')['DataYear']
    .nunique()
    .reset_index(name='YearCount')
)

# Filtra os edif√≠cios para cada caso e exibe apenas as colunas BuildingName e DataYear
display_1_year = df_filtered[df_filtered['BuildingName'].isin(year_counts[year_counts['YearCount'] == 1]['BuildingName'])]
display_2_years = df_filtered[df_filtered['BuildingName'].isin(year_counts[year_counts['YearCount'] == 2]['BuildingName'])]
display_3_years = df_filtered[df_filtered['BuildingName'].isin(year_counts[year_counts['YearCount'] == 3]['BuildingName'])]

# Exibe as tabelas com as colunas desejadas
print("Edif√≠cios com registros em apenas 1 ano:")
print(display_1_year[['BuildingName', 'DataYear']])

print("\nEdif√≠cios com registros em apenas 2 anos:")
print(display_2_years[['BuildingName', 'DataYear']])

print("\nEdif√≠cios com registros em todos os 3 anos (2020, 2021 e 2022):")
print(display_3_years[['BuildingName', 'DataYear']])
```

```python
# Filtrar os dados para os anos de 2020 e 2021
dados_2020 = base_dados[base_dados['DataYear'] == 2021]
dados_2021 = base_dados[base_dados['DataYear'] == 2022]

# Encontrar os edif√≠cios que aparecem em ambos os anos
edificios_ambos_anos = set(dados_2020['BuildingName']).intersection(set(dados_2021['BuildingName']))

# Exibir a lista de edif√≠cios que aparecem em 2020 e 2021
print("Edif√≠cios presentes em 2020 e 2021:")
for edificio in edificios_ambos_anos:
    print(edificio)
# Exibir a quantidade de edif√≠cios que aparecem em ambos os anos
print("\nQuantidade de edif√≠cios presentes em 2020 e 2021:", len(edificios_ambos_anos))
```

```python

# Filtrar os dados para os anos de 2020, 2021 e 2022
dados_2020 = base_dados[base_dados['DataYear'] == 2020]
dados_2021 = base_dados[base_dados['DataYear'] == 2021]
dados_2022 = base_dados[base_dados['DataYear'] == 2022]


# Encontrar os edif√≠cios que aparecem em todos os tr√™s anos
edificios_tres_anos = set(dados_2020['BuildingName']).intersection(set(dados_2021['BuildingName']), set(dados_2022['BuildingName']))

# Encontrar todos os edif√≠cios √∫nicos
todos_edificios = set(base_dados['BuildingName'])

# Encontrar edif√≠cios que n√£o aparecem nos tr√™s anos
edificios_nao_tres_anos = todos_edificios - edificios_tres_anos

# Exibir a lista de edif√≠cios que aparecem em 2020, 2021 e 2022
print("Edif√≠cios presentes em 2020, 2021 e 2022:")
for edificio in edificios_tres_anos:
    print(edificio)

print("\nQuantidade de edif√≠cios presentes em 2020, 2021 e 2022:", len(edificios_tres_anos))

# Exibir a lista de edif√≠cios que N√ÉO aparecem nos tr√™s anos
print("\nEdif√≠cios que N√ÉO aparecem em todos os tr√™s anos (2020, 2021 e 2022):")
for edificio in edificios_nao_tres_anos:
    print(edificio)

print("\nQuantidade de edif√≠cios que N√ÉO aparecem em todos os tr√™s anos:", len(edificios_nao_tres_anos))
```

```python
# Filtrar os dados para os anos de 2019, 2020, 2021 e 2022
dados_2019 = base_dados[base_dados['DataYear'] == 2019]
dados_2020 = base_dados[base_dados['DataYear'] == 2020]
dados_2021 = base_dados[base_dados['DataYear'] == 2021]
dados_2022 = base_dados[base_dados['DataYear'] == 2022]

# Encontrar os edif√≠cios que aparecem em todos os quatro anos
edificios_quatro_anos = set(dados_2019['BuildingName']).intersection(
    set(dados_2020['BuildingName']),
    set(dados_2021['BuildingName']),
    set(dados_2022['BuildingName'])
)

# Encontrar todos os edif√≠cios √∫nicos
todos_edificios = set(base_dados['BuildingName'])

# Encontrar edif√≠cios que n√£o aparecem nos quatro anos
edificios_nao_quatro_anos = todos_edificios - edificios_quatro_anos

# Exibir a lista de edif√≠cios que aparecem de 2019 a 2022
print("Edif√≠cios presentes de 2019 a 2022:")
for edificio in edificios_quatro_anos:
    print(edificio)

print("\nQuantidade de edif√≠cios presentes de 2019 a 2022:", len(edificios_quatro_anos))

# Exibir a lista de edif√≠cios que N√ÉO aparecem nos quatro anos
print("\nEdif√≠cios que N√ÉO aparecem em todos os quatro anos (2019, 2020, 2021 e 2022):")
for edificio in edificios_nao_quatro_anos:
    print(edificio)

print("\nQuantidade de edif√≠cios que N√ÉO aparecem em todos os quatro anos:", len(edificios_nao_quatro_anos))
```
